```python
import pandas as pd
import numpy as np
```

```python
# Задание 1
def traffic_chek(x):
    if x in 'yandex' or x in 'google':
        return 'organic'

    elif x in 'Russia' and ( 'paid' in x  or 'email' in x ) :
        return 'ad'

    elif x not in 'Russia' and ( 'paid' in x  or 'email' in x ) :
        return 'other'
    
    return x.split('_')[-1]
    

df = pd.read_csv('visit_log.csv', sep = ';')

df['source_type'] = df['region'] + '_' + df['traffic_source']
df['source_type'] = df['source_type'].apply(traffic_chek)

df
```

```python
# Задание 3
df = pd.read_csv('ml-latest-small/ratings.csv')
df_min = df.groupby('userId').agg({'movieId': 'count', 'timestamp' : 'min'}).reset_index()
df_max = df.groupby('userId').agg({'movieId': 'count', 'timestamp': 'max'}).reset_index()
df = df_min.merge(df_max, on = 'userId', suffixes=['_min', '_max'])[['userId', 'movieId_max', 	'timestamp_min','timestamp_max']].rename(columns={'movieId_max': 'rat_count'})

np.mean(df[df['rat_count']>= 100]['timestamp_max'] - df[df['rat_count']>= 100]['timestamp_min'])
```
