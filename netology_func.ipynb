```python
import pandas as pd
```

```python
# Задание №1
def rating_class(x):
    if x <= 2:
        return 'Низкий рейтинг'
    elif x > 2 and x <= 4: 
        return 'Средний рейтинг'
    elif x == 4.5 or x == 5 :
        return 'Высокий рейтинг'

df = pd.read_csv('ratings.csv')
df['class'] = df['rating'].apply(rating_class)
df
```

```python
# Задание 2
df = pd.read_csv('keywords.csv')
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

def region_class(x):
    for region, city_lst in geo_data.items():
        for city in city_lst:
            if city in x.lower():
                return region
    return 'undefined'

df['region'] = df['keyword'].apply(region_class)
df['region'].value_counts()

```

```python
# Задание 3
import re
df_rating = pd.read_csv('ratings.csv')
df_movies = pd.read_csv('movies.csv')

df_movies

def chek_year(x):
    res =  re.findall('[1,2][9,0]\d\d', x)
    if len(res) > 0:
        return int(res[-1])
    return 1900
    
df_movies['year'] = df_movies['title'].apply(chek_year)
df_movies = df_movies.merge(df_rating, on = 'movieId', how = 'left')
df_movies.groupby('year').agg({'rating': 'mean'}).reset_index().sort_values('rating', ascending = False)
```
